```python
import matplotlib.pyplot as plt

views = [121,234,454,654,23,45,124]
days = range(1,8)

plt.plot(days,views, color = 'r' , marker = 'o', markerfacecolor = 'g',
                             linestyle = '-.' ,
                             linewidth = '3')
plt.xlabel('no.of days')
plt.ylabel('views')
plt.title('daily views on youtube channel')
plt.show()
```

```python
y_views = [121,232,343,212,543,654,789]
f_views = [232,343,223,543,654,332,564]
t_views = [232,432,431,234,654,787,976]

days = range(1,8)


plt.plot(days , y_views , label = 'Youtube views')
plt.plot(days , f_views , label = 'Facebook views')
plt.plot(days , t_views , label = 'Twiter views')

plt.xlabel('days')
plt.ylabel('views')
plt.legend(loc = 'upper left')
plt.title('Daily views on marketing of individual channels')

```
